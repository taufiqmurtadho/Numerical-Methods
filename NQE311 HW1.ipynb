```python
#This is the code to solve problem no. 1

#This function takes non-negative integer and length of 
#bit sequence as inputs and give the bit sequence of 
#given integer with length l
def int_to_bit(pos_integer, length):
    bit=""
    if pos_integer==0:
        for i in range (length):
            bit+="0"
        return(bit)
    else:
        power=length-1
        for i in range (length):
            if pos_integer-2**(power-i)>=0:
                bit+="1"
                pos_integer-=2**(power-i)
            else:
                bit+="0"
        return(bit)
#this function takes real number and return its floating point 
#representation 1 bit assigned to the sign and the number of bits 
#for mantissa and exponent is to be inputted
def floating_point_rep(real, mantissa, exponent):
    bit=""
    #exception case, real=0
    if real==0:
        for i in range(mantissa+exponent+1):
            bit+="0"
        return(bit)
    #finding sign bit
    elif real>0:
        bit+="0"
    else:
        bit+="1"
        real=-1*real
    #finding exponent bits
    power=-mantissa
    while(real-2**power>=0):
        power+=1
    power-=1
    c=(2**(exponent-1)-1)+power
    bit+=int_to_bit(c,exponent)
    #finding mantissa bits
    a=real-2**power
    for i in range (mantissa):
        if a-2**(power-i-1)>=0:
            bit+="1"
            a-=2**(power-i-1)
        else:
            bit+="0"
    return(bit)

#Evaluation of the functions
print("Machine representation of -52.234375")
print("Single Precision: "+floating_point_rep(-52.234375,23,8))
print("Double Precision: "+floating_point_rep(-52.234375,52,11))
```

```python
#This is a code to solve problem no. 2

#This function takes a bit sequence as an input and return 
#a positive integer
def bit_to_int(bit_seq):
    #Decompose bit string into a list of bit character
    bit_list=list(bit_seq)
    length=len(bit_list)
    
    #Computing the integer
    integer=0
    for i in range(length):
        if bit_list[i]=="1":
            integer+=2**(length-i-1)
    return(integer)

#This function takes a bit sequence in single precision form and 
#convert it into decimals
def bit_to_float(bit_seq):
    #Decompose bit string into a list of bit character
    bit_list=list(bit_seq)
    
    #Separate the sign, exponent, and mantissa parts
    sign_bit=bit_list[0]
    exponent_bit=bit_list[1:9]
    mantissa_bit=bit_list[9:32]

    #Compute the exponent
    exponent=bit_to_int(''.join(exponent_bit))-127
    
    #Compute the mantissa
    if exponent>=0:
        hidden_bit="1"
        integer_part=mantissa_bit[0:exponent]
        decimal_part=mantissa_bit[exponent:32]
        #Compute the integer part
        flt=bit_to_int(hidden_bit+''.join(integer_part))
        #Compute the decimal part
        for i in range(len(decimal_part)):
            if decimal_part[i]=="1":
                flt=flt+2**(-i-1)
    else:
        flt=1
        for i in range(len(mantissa_bit)):
            if mantissa_bit[i]=="1":
                flt=flt+2**(-i-1)
        flt=flt*(2**exponent)
    #decide the sign
    if sign_bit=="0":
        return(flt)
    else:
        return(-flt)
    
#Evaluation of the functions
print("a. "+str(bit_to_float("00000000000000000000000000000000")))
print("b. "+str(bit_to_float("10000000000000000000000000000000")))
print("c. "+str(bit_to_float("01111111100000000000000000000000")))
print("d. "+str(bit_to_float("11111111100000000000000000000000")))
print("e. "+str(bit_to_float("00000000100000000000000000000000")))
print("f. "+str(bit_to_float("01000000101100000000000000000000")))
print("g. "+str(bit_to_float("00111111100000000000000000000000")))
print("h. "+str(bit_to_float("00111101110011001100110011001100")))
```

```python
#This is a code to solve problem no.4
import timeit
#Horner's algorithm

def horner(x):
    c=1
    for i in range(100):
        c=1+c*x
    return(c)

#Naive algorithm
def naive(x):
    c=0
    for i in range(101):
        c+=x**i
    return(c)

#define wrapper function for timing purpose
def wrapper(func, *args, **kwargs):
    def wrapped():
        return func(*args, **kwargs)
    return wrapped

#As a test case, let us consider x=5
print("p(5)= "+str(horner(5)))
print("p(5)= "+str(naive(5)))

#Comparing the execution times for x=5
wrapped1 = wrapper(horner,5)
wrapped2 = wrapper(naive, 5)

#Count the time for 100 executions of each function
print("t(horner) = "+str(timeit.timeit(wrapped1, number=100))+" seconds")
print("t(naive)  = "+str(timeit.timeit(wrapped2, number=100))+" seconds")
```

From the output above, we can see that Horner's algorithm is more efficient becuase it has a much shorter running time. In Horner's algorithm, there are $n$ iterations with one addition and one multiplication in every iteration. Thus, the total number of operation is $2n=200$. Meanwhile, in the naive algorithm, there are n additions and $1+2+3+....+n = \frac{n(n+1)}{2}$ multiplications. Thus, the number of operations in the naive algorithm is $n+\frac{n(n+1)}{2}=5150$ operations.
